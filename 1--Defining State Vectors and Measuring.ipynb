```python
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# Defining the state vector {Basis  (|1>,|0>)}

u = Statevector ([1 / sqrt(2) , 1 / sqrt(2) ])
display(u.draw("latex")) # in nice format
print(u.draw("text"))

# Note that for |00>, use Statevector([1, 0, 0, 0])  for a|00>, use Statevector([a, 0, 0, 0]) and so on..
# (This is via without using tensor yet)

# Measure a qubit {.measure()}...it returns a tuple (measurement_outcome, collapsed_statevector)

outcome, collapsed = u.measure()
print("Outcome:", outcome)
print("Collapsed state:", collapsed)

# Probability

prob = u.probabilities_dict()
print("Probability of outcome:", prob)

# Histogram Visualization

statistics = u.sample_counts(2000)
plot_histogram( statistics )
```
