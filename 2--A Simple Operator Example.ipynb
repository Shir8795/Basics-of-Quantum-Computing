```python
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# Pauli operators

X = Operator([[0, 1],
              [1, 0]])

Y = Operator([[0, -1.0j],
              [1.0j, 0]])

Z = Operator([[1, 0],
              [0, -1]])

# Single-qubit gates

H = Operator([[1/sqrt(2), 1/sqrt(2)],
              [1/sqrt(2), -1/sqrt(2)]])

S = Operator([[1, 0],
              [0, 1.0j]])

T = Operator([[1, 0],
              [0, (1 + 1.0j)/sqrt(2)]])

# initial state |0>

v = Statevector([1, 0])

# evolve through sequence of operators

v = v.evolve(H)
display(v.draw("latex"))
v = v.evolve(T)
display(v.draw("latex"))
v = v.evolve(H)
display(v.draw("latex"))
v = v.evolve(T)
display(v.draw("latex"))
v = v.evolve(Z)
display(v.draw("latex"))
```
