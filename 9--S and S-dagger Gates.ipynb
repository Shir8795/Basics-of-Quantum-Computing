```python
from qiskit .visualization import plot_bloch_vector
from qiskit .visualization import plot_bloch_multivector
from qiskit import QuantumRegister, QuantumCircuit
from numpy import pi
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# =======================
# S Gate (It is /2 rotation around Z axis of Bloch Sphere)
# =======================

S = Operator([[1, 0],
              [0, 1.0j]])

qreg_s = QuantumRegister(1, 'q')
circuit_S = QuantumCircuit(qreg_s)
circuit_S.s(qreg_s[0])

print("S Gate :\n")
display(S)
display(circuit_S.draw(output='mpl'))


sv_in = Statevector.from_label("+")
sv_out = sv_in.evolve(circuit_S)
print("Input - \n")
display(sv_in.draw("bloch"))
display(sv_in.draw("latex"))
print("Output - \n")
display(sv_out.draw("bloch"))
display(sv_out.draw("latex"))


# =======================
# S Dagger Gate (It is /2 rotation around Z axis of Bloch Sphere in the opposite direction)
# =======================

Sd = Operator([[1, 0],
               [0, -1.0j]])

qreg_sd = QuantumRegister(1, 'q')
circuit_Sd = QuantumCircuit(qreg_sd)
circuit_Sd.sdg(qreg_sd[0])   # <-- dagger version in Qiskit

print("S Dagger Gate :\n")
display(Sd)
display(circuit_Sd.draw(output='mpl'))

sv_in1 = Statevector.from_label("+")
sv_out1 = sv_in1.evolve(circuit_Sd)
print("Input - \n")
display(sv_in1.draw("bloch"))
display(sv_in1.draw("latex"))
print("Output - \n")
display(sv_out1.draw("bloch"))
display(sv_out1.draw("latex"))
```
