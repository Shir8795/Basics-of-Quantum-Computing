```python
from qiskit import QuantumRegister, QuantumCircuit
from numpy import pi
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

H = Operator([[1/sqrt(2), 1/sqrt(2)],
              [1/sqrt(2), -1/sqrt(2)]])

# Create 1-qubit quantum register -- it takes 2 arguments 
# a) no of qubits , b) name/label q

# Hadamard Gate

qreg_q = QuantumRegister(1, 'q')

# Create circuit

circuit_1 = QuantumCircuit(qreg_q)

# Pauli-Identity gate denoted by "id" is added

circuit_1.h(qreg_q[0])

# Draw circuit (matplotlib backend)

print("Hadamard Gate :\n")
display(H)
display(circuit_1.draw(output='mpl'))

for basis in ['0', '1']:
    sv_in = Statevector.from_label(basis)
    sv_out = sv_in.evolve(circuit_1)
    print("Input - \n")
    display(sv_in.draw("latex"))
    print("Onput - \n")
    display(sv_out.draw("latex"))




# for 2 qubits

reg1 = QuantumRegister(3,"q")
circuit_2 = QuantumCircuit(reg1)
circuit_2.h(reg1)
display(circuit_2.draw(output='mpl'))

v = Statevector.from_label("000")
v = v.evolve(circuit_2)
display(v.draw("latex"))
```
