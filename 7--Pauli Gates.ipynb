```python
from qiskit import QuantumRegister, QuantumCircuit
from numpy import pi
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

I = Operator([[1, 0],
              [0, 1]])

X = Operator([[0, 1],
              [1, 0]])

Y = Operator([[0, -1.0j],
              [1.0j, 0]])

Z = Operator([[1, 0],
              [0, -1]])

# Create 1-qubit quantum register -- it takes 2 arguments 
# a) no of qubits , b) name/label q

# Pauli I Gate

qreg_q = QuantumRegister(1, 'q')

# Create circuit

circuit_1 = QuantumCircuit(qreg_q)

# Pauli-Identity gate denoted by "id" is added

circuit_1.id(qreg_q[0])

# Draw circuit (matplotlib backend)

print("Pauli I Gate :\n")
display(I)
display(circuit_1.draw(output='mpl'))

for basis in ['0', '1']:
    sv_in = Statevector.from_label(basis)
    sv_out = sv_in.evolve(circuit_1)
    print("Input - \n")
    display(sv_in.draw("latex"))
    print("Onput - \n")
    display(sv_out.draw("latex"))


# Pauli X Gate

qreg_j = QuantumRegister(1, 'q')

# Create circuit

circuit_2 = QuantumCircuit(qreg_j)

# Pauli-Identity gate denoted by "id" is added

circuit_2.x(qreg_q[0])

# Draw circuit (matplotlib backend)

print("Pauli X Gate :\n")
display(X)
display(circuit_2.draw(output='mpl'))

for basis in ['0', '1']:
    sv_in = Statevector.from_label(basis)
    sv_out = sv_in.evolve(circuit_2)
    print("Input - \n")
    display(sv_in.draw("latex"))
    print("Onput - \n")
    display(sv_out.draw("latex"))

# Pauli Y Gate

qreg_l = QuantumRegister(1, 'q')

# Create circuit

circuit_4 = QuantumCircuit(qreg_l)

# Pauli-Identity gate denoted by "id" is added

circuit_4.y(qreg_q[0])

# Draw circuit (matplotlib backend)

print("Pauli Y Gate :\n")
display(Y)
display(circuit_4.draw(output='mpl'))

for basis in ['0', '1']:
    sv_in = Statevector.from_label(basis)
    sv_out = sv_in.evolve(circuit_4)
    print("Input - \n")
    display(sv_in.draw("latex"))
    print("Onput - \n")
    display(sv_out.draw("latex"))


# Pauli Z Gate

qreg_k = QuantumRegister(1, 'q')

# Create circuit

circuit_3 = QuantumCircuit(qreg_k)

# Pauli-Identity gate denoted by "id" is added

circuit_3.z(qreg_q[0])

# Draw circuit (matplotlib backend)

print("Pauli Z Gate :\n")
display(Z)
display(circuit_3.draw(output='mpl'))

for basis in ['0', '1']:
    sv_in = Statevector.from_label(basis)
    sv_out = sv_in.evolve(circuit_3)
    print("Input - \n")
    display(sv_in.draw("latex"))
    print("Onput - \n")
    display(sv_out.draw("latex"))


```
