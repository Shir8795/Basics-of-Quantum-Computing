```python
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# Creating statevectors using label

zero = Statevector.from_label("0")
display(zero.draw("latex"))
one = Statevector.from_label("1")
display(one.draw("latex"))
plus = Statevector.from_label("+")
display(plus.draw("latex"))
minus = Statevector.from_label("-")
display(minus.draw("latex"))

# Tensor Product of statevectors

tensor1 = plus.tensor(minus)
display(tensor1.draw("latex"))

# Tensor Product of Operators

X = Operator([[0, 1] , [1, 0]])
I = Operator([[1, 0] , [0, 1]])
G = X.tensor(I)
display(G)

# Tensor Product of Operators on Tensor Product of statevectors

tensor1.evolve(G).draw("Latex")
```
