```python
from qiskit .visualization import plot_bloch_vector
from qiskit .visualization import plot_bloch_multivector
from qiskit import QuantumRegister, QuantumCircuit
from numpy import pi
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# =======================
# T Gate (It is /4 rotation around Z axis of Bloch Sphere)
# =======================

T = Operator([[1, 0],
              [0, (1.0 + 1.0j)/sqrt(2)]])   # e^{i/4} = (1 + i)/2

qreg_t = QuantumRegister(1, 'q')
circuit_T = QuantumCircuit(qreg_t)
circuit_T.t(qreg_t[0])

print("T Gate :\n")
display(T)
display(circuit_T.draw(output='mpl'))

sv_in_t = Statevector.from_label("+")   # |+> = (|0> + |1>)/2
sv_out_t = sv_in_t.evolve(circuit_T)

print("Input - \n")
display(sv_in_t.draw("bloch"))
display(sv_in_t.draw("latex"))

print("Output - \n")
display(sv_out_t.draw("bloch"))
display(sv_out_t.draw("latex"))


# =======================
# T Dagger Gate (It is -/4 rotation around Z axis of Bloch Sphere)
# =======================

Td = Operator([[1, 0],
               [0, (1.0 - 1.0j)/sqrt(2)]])   # e^{-i/4} = (1 - i)/2

qreg_td = QuantumRegister(1, 'q')
circuit_Td = QuantumCircuit(qreg_td)
circuit_Td.tdg(qreg_td[0])   # <-- dagger version in Qiskit

print("T Dagger Gate :\n")
display(Td)
display(circuit_Td.draw(output='mpl'))

sv_in_td = Statevector.from_label("+")
sv_out_td = sv_in_td.evolve(circuit_Td)

print("Input - \n")
display(sv_in_td.draw("bloch"))
display(sv_in_td.draw("latex"))

print("Output - \n")
display(sv_out_td.draw("bloch"))
display(sv_out_td.draw("latex"))
```
