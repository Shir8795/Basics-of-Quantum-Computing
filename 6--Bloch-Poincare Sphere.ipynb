```python
# for interactive
%matplotlib widget 

# for static images use %matplotlib inline

from qiskit .visualization import plot_bloch_vector
from qiskit .visualization import plot_bloch_multivector
from qiskit.quantum_info import *
from qiskit . visualization import plot_histogram
import numpy as np
from IPython.display import Math, display

# Take theta and phi in degrees from user as input

theta_deg = float(input("Enter  in degrees: "))
phi_deg   = float(input("Enter  in degrees: "))
theta = np.deg2rad(theta_deg)
phi   = np.deg2rad(phi_deg)

# Build the state |> = cos(/2)|0> + e^{i} sin(/2)|1>

alpha = np.cos(theta / 2)
beta  = np.exp(1j * phi) * np.sin(theta / 2)
psi = Statevector([alpha, beta])
display(psi.draw("latex"))

# Plotting the bloch vector

x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)
print(f"Bloch vector = (x={x:.3f}, y={y:.3f}, z={z:.3f})")

plot_bloch_vector([x, y, z], title='Bloch Sphere')
```
