```python
# Import Qiskit tools
from qiskit import QuantumCircuit
from qiskit.quantum_info import *
from qiskit.visualization import plot_histogram
from numpy import sqrt
from IPython.display import Math, display

# Create a 1-qubit circuit

circuit = QuantumCircuit(1)

# Adding quantum gates
# In bracket 0 implies i am applying hadamard to the zeroth qubit

circuit.h(0)    # Hadamard gate: |0> → (|0> + |1>)/√2
circuit.t(0)    # T gate: adds a π/4 phase to |1>
circuit.h(0)    # Apply Hadamard again
circuit.t(0)    # Another phase gate
circuit.z(0)    # Pauli-Z gate: flips phase of |1>


# Draw the circuit and use "display" in front of it always to show

display(circuit.draw(output='mpl', style='iqp'))     


# make a qubit

qb1=Statevector([1, 0])
display(qb1.draw("latex"))
qb1=qb1.evolve(circuit)
display(qb1.draw("latex"))

#Visualization and Measurement

statistics = qb1.sample_counts(4000)
plot_histogram( statistics )

```
